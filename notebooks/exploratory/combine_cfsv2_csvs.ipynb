```python
import pandas as pd
import os

# Path to the parent folder containing the year folders
parent_folder_path = 'C:/Users/fitzpatrick/Desktop/Data/CFS'

# List to hold the dataframes
df_list = []

# Iterate over each year folder (assuming the folders are named by year, e.g., '2020', '2021', ...)
for year_folder in os.listdir(parent_folder_path):
    year_folder_path = os.path.join(parent_folder_path, year_folder)
    
    # Check if the folder is a directory (and likely contains CSV files)
    if os.path.isdir(year_folder_path):
        # Construct the full path to the CSV file (assuming only one file per year folder)
        csv_file_path = os.path.join(year_folder_path, 'CFS_TMP_Basin_Avgs.csv')
        
        if os.path.exists(csv_file_path):  # Ensure the file exists
            # Read the CSV into a dataframe
            df = pd.read_csv(csv_file_path)
            
            # Add the year column to the dataframe
            #df['year'] = year_folder  # Use the folder name (year) as the year
            
            # Append the dataframe to the list
            df_list.append(df)

# Concatenate all dataframes into one
combined_df = pd.concat(df_list, ignore_index=True)

# Show the combined dataframe
print(combined_df)

# Optionally, save the combined dataframe to a new CSV file
combined_df.to_csv(parent_folder_path + '/CFS_TMP_Basin_Avgs.csv', index=False)
```
