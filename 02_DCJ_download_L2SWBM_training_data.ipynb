```python
import pandas as pd
import requests
import zipfile
import io
import os
```

```python
# Function to download and extract the zip file
def download_and_unzip(url, extract_to='L2SWBM_training_data'):
    try:
        r = requests.get(url)
        # Check if the request was successful
        if r.status_code == 200:
            z = zipfile.ZipFile(io.BytesIO(r.content))
            z.extractall(extract_to)
            print(f"Data extracted to '{extract_to}' directory")
        else:
            print(f"Error: Unable to download the file. HTTP status code {r.status_code}")
    except requests.RequestException as e:
        print(f"Error: An issue occurred while downloading the data.\n{e}")
    except zipfile.BadZipFile as e:
        print(f"Error: The downloaded file is not a valid zip file.\n{e}")
```

```python
# URL to the data zip file
data_url = 'https://deepblue.lib.umich.edu/data/downloads/np193931r'

# Use the function to download and unzip the data
download_and_unzip(data_url)

# Path to the directory where files are extracted
data_directory = 'L2SWBM_training_data/ts_posterior'
```

```python
# Check if the path exists and is a directory
if os.path.exists(data_directory) and os.path.isdir(data_directory):
    # Dictionary to hold data frames
    data_frames = {}

    try:
        # Loop through all files in the directory
        for filename in os.listdir(data_directory):
            # Check if the file is a CSV
            if filename.endswith('.csv'):
                file_path = os.path.join(data_directory, filename)
                try:
                    data_frames[filename] = pd.read_csv(file_path)
                    #print(f"Loaded {filename} into DataFrame:")
                    #print(data_frames[filename].head(), '\n')
                except pd.errors.EmptyDataError as e:
                    print(f"Warning: The file '{filename}' is empty. Skipping.")
                except pd.errors.ParserError as e:
                    print(f"Error: An error occurred while parsing '{filename}'.\n{e}")
                except Exception as e:
                    print(f"Error: An unexpected error occurred while reading '{filename}'.\n{e}")
    except FileNotFoundError as e:
        print(f"Error: The directory {data_directory} does not exist.\n{e}")
else:
    print(f"Error: The specified path '{data_directory}' does not exist or is not a directory.")

# At this point, data_frames contains all your CSVs
# You can access each DataFrame using its corresponding filename as the key, for example:
# df = data_frames['example_filename.csv']
```
